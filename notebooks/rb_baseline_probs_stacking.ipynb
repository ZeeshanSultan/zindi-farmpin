# Probabilities Stacking

Mean-stack the output probabilities of the current best model for each timestamp to create new predictions

```python
import os
import numpy as np
import pandas as pd
from glob import glob

from config import interim_data_dir
```

## Mean-stack all test probabilities

```python
files_list = glob(os.path.join(interim_data_dir, 'baseline_probs/test/*.csv'))

data_list = []
for fp in files_list:
    probs_df = pd.read_csv(fp, index_col=0)
    
    data_list.append(probs_df.values)

data = np.stack(data_list)

mean_df = pd.DataFrame(data.mean(axis=0), columns = probs_df.columns)
mean_df.index = probs_df.index.values

mean_df.head()
```

```python
mean_df.to_csv('../../data/submissions/baseline_tpot_mean_stacked.csv')
```

> Moved up to #1 !

## Median Stacking

```python
median_df = pd.DataFrame(np.median(data, axis=0), columns=probs_df.columns)
median_df.index = probs_df.index.values
median_df.to_csv('../../data/submissions/baseline_tpot_median_stacked.csv')
median_df.head()
```

> Similar to mean stacking but not as good...
